```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
nsw = pd.read_csv('./PRICE_AND_DEMAND_NSW.csv', index_col=0, )
nsw
```

```python
std, mean = nsw.DEMAND.values.std(), nsw.DEMAND.values.mean()
std, mean
```

```python
plt.figure(figsize=(20,5))
plt.plot((nsw.DEMAND.values-mean)/std)
```
